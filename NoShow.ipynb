# 4. Data Cleaning and Processing
In this section, we will identify missing values, and duplicate values. We will also check for inconsistent value. We will also check misspelled and outlier data.

ScheduledDay and AppointmentDay are currently objects, it will be converted to DateTime
AppointmentDay's time will be dropped (as it is set as 00:00:00)
Misspelled columns are going to be renamed
Erroneous data from the Age column will be deleted

## 4.2 Duplicate Values

```python
# Finding Duplicates
duplicate_rows = sum(base_data.duplicated())
print("Number of duplicated rows : ",duplicate_rows)
duplicate_patientid =base_data['PatientId'][base_data['PatientId'].duplicated() == True].count()
print("Number of duplicated patient  : ",duplicate_patientid)
```

This data set does not have duplicate appointments but has 48,228 patients that can be considered as returning patients.

## 4.3 Data type conversions:
Ensure that each column is in the appropriate data type. Convert columns to their correct data types (e.g., converting dates to datetime objects) for better analysis and modeling.

```python
# #modifying the date and time into standard form
base_data['ScheduledDay'] = pd.to_datetime(base_data['ScheduledDay']).dt.date.astype('datetime64[ns]')
# Convert ScheduledDay from 'object' type to 'datetime'
base_data['AppointmentDay'] = pd.to_datetime(base_data['AppointmentDay']).dt.date.astype('datetime64[ns]')
```

```python
# Data cleaning steps
# dropping some columns which have no significance
#base_data.drop(['PatientId', 'AppointmentID', 'Neighbourhood'], axis=1, inplace=True)
```

```python
# base_data.head()
```

## 4.4 Rename misspelled column names:

```python
# Renaming misspelled column names
base_data= base_data.rename(columns={'Hipertension': 'Hypertension', 'Handcap': 'Handicap', 'SMS_received': 'SMSReceived'})
```

```python
# change all the Handicap set above 0 as 1
base_data.loc[base_data.Handicap > 0, 'Handicap'] = 1
# check
print('Values in Handicap:',base_data.Handicap.unique())
```

```python
# Convert Integer to Boolean
# Columns Scholarship, Hypertension, Diabetes, Alcoholism, Handicap, SMS_received should be binary data set as 0 for False and 1 for True
# Scholarship
base_data['Scholarship'] = base_data['Scholarship'].apply(lambda x: True if x==1 else False)
# Hypertension
base_data['Hypertension'] = base_data['Hypertension'].apply(lambda x: True if x==1 else False)
# Diabetes
base_data['Diabetes'] = base_data['Diabetes'].apply(lambda x: True if x==1 else False)
# Handicap
base_data['Handicap'] = base_data['Handicap'].apply(lambda x: True if x==1 else False)
# Alcoholism
base_data['Alcoholism'] = base_data['Alcoholism'].apply(lambda x: True if x==1 else False)
# SMS_received
base_data['SMSReceived'] = base_data['SMSReceived'].apply(lambda x: True if x==1 else False)
```

Rename “No-show” column to “Presence” and its values to ‘Present’ and ‘Absent’ so as to avoid any misinterpretation.

```python
base_data['Presence'] = base_data['No-show'].apply(lambda x: 'Present' if x == "No" else 'Absent')
base_data = base_data.drop('No-show',axis=1)
```

```python
base_data.columns
```

## 4.5 Identify the outlier 

```python
# Look at the distribution of the Age feature
plt.figure(figsize=(16,2))
plt.xticks(rotation=90)
dist_age = sns.boxplot(x=base_data.Age)
dist_age.set_title("Patients distribution by Age")
plt.show()
```

Most the patients are between 18 and 55 years old. It is clearly shown that the value 115 and -1 in age columns are outlier.So we will eliminate the patients records who are aged 115 years and -1 year.

```python
# delete patients that are over 100
base_data = base_data.drop(base_data[base_data.Age > 100].index)
# delete the patients that are negative
base_data = base_data.drop(base_data[base_data.Age < 0].index)
```

## 4.6 Feature engineering
We can add a new feature to the dataset — ‘Waiting Time Days’ to check how long the patient needs to wait for the appointment day.Another new feature may be ‘WeekDay’ — a weekday of an appointment. With this feature, we can analyze on which days people don’t show up more often.Similarly, add ‘Month’, ‘Hour’ features.

For the schedule day and appointment day storing the weekdays only into a variable

```python
# base_data['sch_weekday'] = base_data['ScheduledDay'].dt.dayofweek
# base_data['app_weekday'] = base_data['AppointmentDay'].dt.dayofweek
# base_data['sch_weekday'].value_counts()
base_data['WeekDay'] = base_data['AppointmentDay'].apply(lambda x: x.weekday())
replace_map = {'WeekDay': {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday'}}
base_data.replace(replace_map, inplace=True)
#base_data.head()
base_data['SchWeekDay'] = base_data['ScheduledDay'].apply(lambda x: x.weekday())
replace_map2 = {'SchWeekDay': {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday'}}
base_data.replace(replace_map2, inplace=True)
base_data.head()
```

```python
base_data['Waiting_Time_Days'] = base_data['AppointmentDay']-base_data['ScheduledDay']
base_data['Waiting_Time_Days'] = base_data['Waiting_Time_Days'].dt.days
base_data.head()
# base_data['Waiting Time Days'].max()
```

```python
base_data.describe()
# Describe only works for numerical value.
```

```python
# Look at the distribution of the Waiting Time Days
plt.figure(figsize=(16,2))
plt.xticks(rotation=90)
WTime = sns.boxplot(x=base_data.Waiting_Time_Days)
WTime.set_title("Waiting Time by Days")
plt.show()
```

Similarly, add ‘Month’, ‘Hour’ features:

```python
base_data['Month'] = base_data['AppointmentDay'].dt.month_name()
base_data['Hour'] = base_data['ScheduledDay'].dt.hour
base_data.head()
#base_data.describe()
```

### Waiting Time Between Booking and Medical Appointment


```python
# Look at the distribution of Waiting_Days
plt.figure(figsize=(16,2))
plt.xticks(rotation=90)
dist_age = sns.boxplot(x=base_data.Waiting_Time_Days)
dist_age.set_title("Patients distribution by Waiting Time")
plt.show()
```

```python
# create pre-appointment period groups using the distribution above
bins= [0,1,7,14,21,28,35,200]
labels = ['Same day','Less than a week','A week','Two weeks','Three weeks','A month','More than a month']
base_data['Pre_Appt_Period'] = pd.cut(base_data['Waiting_Time_Days'], bins=bins, labels=labels, right=False)
# replace missing and erronous values with unknown
base_data['Pre_Appt_Period'] = base_data['Pre_Appt_Period'].cat.add_categories('unknown').fillna('unknown')

base_data.groupby(['Pre_Appt_Period'])['PatientId'].count()
```

```python
# barchart showing number of show/ no-show in percentage by the amount of time between booking and appointment
my_colors = sns.color_palette("Paired") 
psla = base_data.groupby('Pre_Appt_Period')['Presence'].value_counts(normalize=True).unstack().plot(kind='bar',color=my_colors, figsize=(15,5))
psla.set_title("Percentage of Appointment Show/No Show by Time Between Booking and Appointment")
plt.show()
```

We can clearly identify that people missed their appointment because of the long waiting time between appointment booking and appointment day.
